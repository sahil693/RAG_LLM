```python
#fetch the pdf from url
import requests
url="https://courses.cs.umbc.edu/671/fall12/notes/python/04_python_functions.pdf"
response=requests.get(url)
response.raise_for_status()
```

```python
#save the pdf local file
with open("python_functions.pdf","wb")as f:
  f.write(response.content)
```

```python
with open("python_functions.pdf","rb")as f:
  loader=UnstructuredFileIOLoader(file=f)
  documents=loader.load()

documents
```

```python
text_splitter=CharacterTextSplitter(chunk_size=200,chunk_overlap=50)
text_splitter
```

```python
texts=text_splitter.split_documents(documents)
texts
```

```python
type(texts)
```

```python
texts[2]
```

```python
embeddings=HuggingFaceEmbeddings()
```

```python
persist_directory="vector_db"
```

```python
vectordb=Chroma.from_documents(texts,embeddings,persist_directory=persist_directory)
```

```python
#retrieval
retriever=vectordb.as_retriever()
```

```python
llm=ChatGroq(model="Llama-3.1-70b-Versatile",temperature=0)
```

```python
qa_chain=RetrievalQA.from_chain_type(llm=llm,chain_type="stuff",retriever=retriever)
```

```python
query="what are the built-in function form this pdf"
response=qa_chain.invoke({"query":query})
```
